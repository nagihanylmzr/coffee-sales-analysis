```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(
    x="coffee_name",
    data=df,
    order=df["coffee_name"].value_counts().index,
    palette="husl"
)

plt.title("coffee_name Dağılımı", fontsize=14)
plt.xticks(rotation=45)

# 200 seviyesine kırmızı çizgi ekleme
plt.axhline(200, color="red", linestyle="--", linewidth=1.5, label="200 sınırı")

# Legend ekleme
plt.legend()

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Tema
sns.set_theme(style="whitegrid")

# Grafik alanı
f, ax = plt.subplots(figsize=(7, 6))
sns.despine(f, left=True, bottom=True)

# Scatter plot
sns.scatterplot(
    x="hour_of_day",
    y="money",
    hue="coffee_name",         # Kahve türüne göre renk
    size="money",              # Harcanan paraya göre nokta boyutu
    palette="ch:r=-.2,d=.3_r", # Renk paleti
    sizes=(20, 200),           # Nokta boyutu aralığı
    linewidth=0,
    data=df,
    ax=ax
)

plt.title("Saatlere Göre Harcama ve Kahve Türleri")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="dark")

# Scatter + Histogram + KDE (yoğunluk konturları)
f, ax = plt.subplots(figsize=(6, 6))

sns.scatterplot(
    x="hour_of_day",
    y="money",
    s=5, color=".15",
    data=df
)

sns.histplot(
    x="hour_of_day",
    y="money",
    bins=50,
    pthresh=.1,
    cmap="mako",
    data=df
)

sns.kdeplot(
    x="hour_of_day",
    y="money",
    levels=5,
    color="w",
    linewidths=1,
    data=df
)

plt.title("Saat ve Harcama İlişkisi - Yoğunluk Grafiği")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="white")

# JointGrid: Scatter + KDE + Histogram
g = sns.JointGrid(data=df, x="hour_of_day", y="money", space=0)

# Ortak grafik (KDE yoğunluk)
g.plot_joint(
    sns.kdeplot,
    fill=True,
    clip=((df["hour_of_day"].min(), df["hour_of_day"].max()),
          (df["money"].min(), df["money"].max())),
    thresh=0,
    levels=100,
    cmap="rocket"
)

# Kenar histogramları
g.plot_marginals(sns.histplot, color="#03051A", alpha=1, bins=25)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.set_theme(style="white")

# Matplotlib figure ayarı
f, axes = plt.subplots(3, 3, figsize=(9, 9), sharex=True, sharey=True)

# Farklı cubehelix paletleri ile KDE plot
for ax, s in zip(axes.flat, np.linspace(0, 3, 9)):

    # Cubehelix colormap
    cmap = sns.cubehelix_palette(start=s, light=1, as_cmap=True)

    # hour_of_day & money KDE plot
    sns.kdeplot(
        x="hour_of_day",
        y="money",
        data=df,
        cmap=cmap, fill=True,
        cut=10, thresh=0, levels=15,
        ax=ax,
    )
    ax.set_axis_off()

# Limitler: verinin min/max değerlerine göre ayarlanıyor
ax.set(
    xlim=(df["hour_of_day"].min(), df["hour_of_day"].max()),
    ylim=(df["money"].min(), df["money"].max())
)

f.subplots_adjust(0, 0, 1, 1, .08, .08)
plt.show()
```
